```python
# import CSV file
df_adult = pd.read_csv("adult.csv")
df_adult.head(10)
```

```python
# 1.cleaning data
df_adult = df_adult[~df_adult.apply(lambda row: row.astype(str).str.contains(r'\?').any(), axis=1)]
```

```python
df_adult.info()
```

```python
# 1.2 checking data
mask_valid = (
    ((df_adult["salary"] == "<50K") & (df_adult["salary K$"] <= 50)) |
    ((df_adult["salary"] == ">50K") & (df_adult["salary K$"] >= 50))
              )
invalid_rows = df_adult[~mask_valid]
```

```python
# 1.3 changing header
df_adult = df_adult.rename(columns=
                           {"marital-status": "marital_status",
                            "hours-per-week": "hours_per_week",
                            "native-country": "native_country",
                            }
                           )
print(df_adult.columns)
```

```python
# 1.4 reâ€‘indexing the DataFrame
df_adult = df_adult.drop(columns=["id"], errors="ignore")
df_adult = df_adult.set_index("Unnamed: 0")
df_adult.index = range(1, len(df_adult) + 1)
print(df_adult.head())
```

```python
df_adult.head()
```

```python
# 2.Count m and f
df_adult["sex"].value_counts()
```

```python
# 3.avg m age
avg_male = df_adult[df_adult['sex'] == 'Male']['age'].mean()
print(avg_male)
```

```python
# 4. Percentage people of Poland
poland_people = df_adult[df_adult["native_country"] == "Poland"].shape[0]
print(poland_people)
total_people = df_adult.shape[0]
percentage_poland = (poland_people / total_people) * 100
print(f'Percentage of people from Poland: {percentage_poland:.2f}%')
```

```python
# 5. People without education but > 50K
higher_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
w_education = df_adult[(~df_adult["education"].isin(higher_education)) & (df_adult["salary"] == '>50K')]
print("ÐšÑ–Ð»ÑŒÐºÑ–ÑÑ‚ÑŒ Ð»ÑŽÐ´ÐµÐ¹ Ð±ÐµÐ· Ð²Ð¸Ñ‰Ð¾Ñ— Ð¾ÑÐ²Ñ–Ñ‚Ð¸ Ð· >50K:", w_education.shape[0])
total_rich = df_adult[df_adult["salary"] == '>50K'].shape[0]
share = w_education.shape[0] / total_rich * 100
print(f"Ð§Ð°ÑÑ‚ÐºÐ° Ð±ÐµÐ· Ð²Ð¸Ñ‰Ð¾Ñ— Ð¾ÑÐ²Ñ–Ñ‚Ð¸ ÑÐµÑ€ÐµÐ´ >50K: {share:.2f}%")
```

```python
# 6. Age statistics for each education
grouped_by_age_education = df_adult.groupby("education")["age"].describe()
print(grouped_by_age_education)
```

```python
# 7. Compare salaries of married and unmarried men
male_df = df_adult[df_adult["sex"] == "Male"].copy()
male_df["married"] = male_df["marital_status"].str.startswith("Married")
counts = male_df[["married", "salary"]].value_counts().reset_index(name="count")
print(counts)
```

```python
# 8. Get the maximum number of hours a person can work per week
max_hours = df_adult["hours_per_week"].max()
same_hours = df_adult["hours_per_week"] == max_hours
same_hours_count = same_hours.sum()
print(same_hours_count)
```
